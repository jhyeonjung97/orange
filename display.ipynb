```python
from ase.visualize import view
from ase.io import read

atoms=read('POSCAR')
view(atoms)
```

```python
view(atoms, viewer='x3d')
```

```python
view(atoms, viewer='ngl')
```

```python
from ase.io import write
write('image.png', atoms)
```
